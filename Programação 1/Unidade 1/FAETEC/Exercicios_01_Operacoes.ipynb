###Questão 1

Crie uma função que contenha uma estrutura condicional que imprima na tela se um número é positivo ou negativo. Além disso, imprima um aviso de que não é possível testar o valor `0`.

````python
# Questao 1

# Funcao que verifica se um numero e positivo ou negativo

n1 = int(input("Digite um número e direi se é positivo ou negativo: "))

if n1>0:
  tipo = "positivo"

elif n1==0:
  tipo = "neutro, não sendo possível determinar valor positivo ou negativo";

else:
  tipo = 'negativo'

print(f'Seu número digitado foi {n1} e esse número é {tipo}')
````
