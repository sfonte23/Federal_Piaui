###Questão 3

Abaixo você encontra uma funcao que calcula o triplo de um valor. Embora, completamente sem documentação. Complete a documentação utilizando comentário iniciados com $\#$.

```python
# Questao 3

#############################
# Documente a funcao abaixo:
#############################

def triplo(numero):
    """Funcao que calcula o triplo de um numero.

    Esta função recebe um número como entrada e retorna o valor triplicado desse número.

    Parâmetros
    ----------
    numero : float
        O número para o qual o triplo será calculado.

    Retorna
    -------
    resultado : float
        O valor do número multiplicado por 3 (triplo do número).

    Exemplo
    -------
    >>> triplo(5)
    15.0
    """

    # Calcula o triplo do número fornecido
    resultado = 3 * numero

    # Retorna o resultado calculado
    return resultado

# Exemplo de uso
triplo_do_numero = triplo(5)
print('Triplo:', triplo_do_numero)
```
