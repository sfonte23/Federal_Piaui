### Removendo dados faltantes

üëª Dados faltantes s√£o valores ausentes em um conjunto de dados. Eles podem ocorrer por uma variedade de raz√µes, incluindo erros de entrada, perda de dados ou aus√™ncia de informa√ß√µes.

> üõ†Ô∏è *Exemplo pr√°tico 1:* Obtendo s√©ries s√≥cioec√¥micas do Brasil e removendo valores faltantes (Fonte: Banco Mundial)
*   Expectativa de vida ao nascer - Brasil: https://data.worldbank.org/indicator/SP.DYN.LE00.IN?locations=BR
*   Taxa de desemprego entre jovens - Brasil: https://data.worldbank.org/indicator/EN.ATM.CO2E.PC?locations=BR
*   Acesso √† eletricidade - Brasil: https://data.worldbank.org/indicator/EG.ELC.ACCS.ZS?locations=BR
*   Preval√™ncia de desnutri√ß√£o - Brasil: https://data.worldbank.org/indicator/SN.ITK.DEFC.ZS?locations=BR

```python
# Carregando pacote `Pandas`
import pandas as pd
```

```python
# Carregando m√≥dulo (parte do pacote)
from pandas_datareader import wb
```

```python
# Definindo c√≥digo
codigo_wb = ['SP.DYN.LE00.IN',   # Expectativa de vida ao nascer (anos)
             'EN.ATM.CO2E.PC',   # Emiss√£o de CO2 (toneladas per capita)
             'EG.ELC.ACCS.ZS',   # Acesso √† eletricidade (% da popula√ß√£o)
             'SN.ITK.DEFC.ZS']   # Preval√™ncia de desnutri√ß√£o (% da popula√ß√£o)

# Obtendo dados
df_wb = wb.download(indicator = codigo_wb,  # Lista de indicadores
                    country = ['BR'],       # Lista de pa√≠ses
                    start = 2014,           # Data in√≠cio AAAA
                    end = 2023)             # Data fim AAAA

# Resultado (tabela)
df_wb
```

üí¨ Note que a lista deste indicadores possuem valores faltantes `NaN`. Em um relat√≥rio gerencial, iremos disponibilizar somente os dados completos da seguinte forma:

```python
# Removendo dados faltantes com o
relatorio = df_wb.dropna()

# Resultado (tabela sem valores faltantes)
relatorio
```

> ‚ö†Ô∏è As linhas dos anos 2021 e 2022 foram exclu√≠das!

> ‚ö†Ô∏è Toda a linha √© exclu√≠da, incluindo as colunas que possuem informa√ß√£o (como na linha do ano de 2021)!

> üó£Ô∏è Voc√™ pode substituir um `DataFrame` por ele mesmo. Aqui estamos dando nomes diferentes por fins educativos. Voc√™ pode o sobrepor dando o mesmo o nome, por exemplo: `df_wb = df_wb.dropna()`

* Na linha mais alta, temos: `SP.DYN.LE00.IN`,	`EN.ATM.CO2E.PC`,	`EG.ELC.ACCS.ZS` e	`SN.ITK.DEFC.ZS`

### Criando colunas

> üõ†Ô∏è *Exemplo pr√°tico 5:* A√ß√µes americanas em valores em reais (R\$)
*   Para um melhor entendimento dos investidores, vamos converter os valores dessas a√ß√µes em d√≥lares americanos (US\$) para reais (R$).

```python
# Carregando pacote
import yfinance as yf
```

```python
# Definindo abrevia√ß√£o (ticker)
ticker = ["MSFT",  # Microsoft
          "AAPL"]  # Apple

# Obtendo dados
df_acoesEUA = yf.download(ticker,                # C√≥digo das a√ß√µes
                          start="2023-07-01",    # Data inicial AAAA-MM-DD
                          end="2023-07-14")      # Data final AAAA-MM-DD

# Resultado (tabela)
df_acoesEUA
```

üí¨ Note que, agora, o √≠ndice das colunas tem 2 n√≠veis. Por outro lado, o √≠ndice das linhas tem somente 1.

üîÑ Caso contr√°rio do exemplo anterior.

```python
# Visualizando os √≠ndices de linha
df_acoesEUA.index
```

```python
# Visualizando os √≠ndices de linha
df_acoesEUA.columns
```

üßπ Vamos organizar este `DataFrame` com o que aprendemos at√© agora:
1.    Somente os valores *Adj Close*`*` (fechamento ajustado) nos interessam.
2.    Queremos que a data seja uma coluna (e n√£o o √≠ndice de linha)
3.    Vamos renomear as colunas

> `*`üìñ O fechamento ajustado √© o pre√ßo de fechamento de uma a√ß√£o ap√≥s os ajustes para todas as distribui√ß√µes de dividendos e a√ß√µes corporativas que ocorreram em qualquer momento antes da abertura do dia seguinte.

```python
# TAREFA 1: Somente os valores *Adj Close* nos interessam

# Selecionando somente as 2 primeiras colunas (Adj Close)
relat_acoes = df_acoesEUA.iloc[:, [0, 1]]
```

```python
# TAREFA 2: Queremos que a data seja uma coluna

# Reiniciando √≠ndice
relat_acoes = relat_acoes.reset_index()
```

```python
# TAREFA 3: Vamos renomear as colunas

# Renomeando colunas
relat_acoes.columns = ['Data', 'Apple', 'Microsoft']
```

```python
# Visualizando o relat√≥rio
relat_acoes
```

üëâ Agora, vamos criar uma nova coluna com a cota√ß√£o das a√ß√µes em reais (R$):[texto do link](https://)

```python
# Cota√ß√£o do D√≥lar para Real
cotacao = 5.43   # > Ou seja, US$ 1,00 vale R$ 5,43 (suposi√ß√£o)

# Criando novas colunas
relat_acoes["Apple_Reais"] = relat_acoes["Apple"] * cotacao
relat_acoes["Microsoft_Reais"] = relat_acoes["Microsoft"] * cotacao
```

```python
# Visualizando o relat√≥rio
relat_acoes
```

### Filtrando colunas
