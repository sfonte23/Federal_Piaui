## Finalidades e Objetivos do Manuseio de Dados Financeiros:

1.	**Análise de Tendências**: Os dados históricos permitem identificar padrões e tendências que podem influenciar o comportamento futuro do mercado. Isso é fundamental para os investidores anteciparem movimentos de preços e tomarem decisões de compra ou venda com base em evidências sólidas.

2.	**Previsão e Modelagem**: A análise estatística e matemática de dados financeiros possibilita a criação de modelos preditivos. Esses modelos podem ser usados para prever possíveis cenários de mercado e avaliar o impacto de diferentes variáveis econômicas.

3.	**Gestão de Riscos**: A identificação e avaliação de riscos são fundamentais no mercado financeiro. O manuseio de dados permite calcular métricas de risco, como o *VaR*, para entender a exposição de um investimento a movimentos adversos do mercado.

4.	**Alocação de Ativos**: Os dados auxiliam na tomada de decisões sobre a alocação de ativos, ajudando a diversificar portfólios e a equilibrar riscos e retornos de acordo com os objetivos do investidor.

5.	**Avaliação de Desempenho**: O acompanhamento contínuo do desempenho dos investimentos é possível graças ao manuseio de dados. Isso permite avaliar a eficácia das estratégias adotadas e fazer ajustes conforme necessário.

6.	**Identificação de Oportunidades**: A análise de dados pode revelar oportunidades de investimento que podem não ser óbvias à primeira vista. Isso é especialmente valioso em um ambiente de mercado competitivo.

7.	**Tomada de Decisões Informadas**: Ao fornecer *insights* baseados em dados, o manuseio de dados capacita os profissionais financeiros a tomar decisões mais informadas e confiantes, reduzindo a influência de emoções e intuições no processo de tomada de decisões.

> 🎲⛏️ O manuseio de dados no mercado financeiro é uma ferramenta poderosa para compreender, analisar e reagir às complexidades em constante evolução do mundo financeiro. Seja para previsões de mercado, gestão de riscos ou alocação de ativos, o uso eficaz de dados é uma competência essencial para os participantes desse ambiente dinâmico e desafiador.

### Removendo dados faltantes

👻 Dados faltantes são valores ausentes em um conjunto de dados. Eles podem ocorrer por uma variedade de razões, incluindo erros de entrada, perda de dados ou ausência de informações.

> 🛠️ *Exemplo prático 1:* Obtendo séries sócioecômicas do Brasil e removendo valores faltantes (Fonte: Banco Mundial)
*   Expectativa de vida ao nascer - Brasil: https://data.worldbank.org/indicator/SP.DYN.LE00.IN?locations=BR
*   Taxa de desemprego entre jovens - Brasil: https://data.worldbank.org/indicator/EN.ATM.CO2E.PC?locations=BR
*   Acesso à eletricidade - Brasil: https://data.worldbank.org/indicator/EG.ELC.ACCS.ZS?locations=BR
*   Prevalência de desnutrição - Brasil: https://data.worldbank.org/indicator/SN.ITK.DEFC.ZS?locations=BR

```python
# Carregando pacote `Pandas`
import pandas as pd
```

```python
# Carregando módulo (parte do pacote)
from pandas_datareader import wb
```

```python
# Definindo código
codigo_wb = ['SP.DYN.LE00.IN',   # Expectativa de vida ao nascer (anos)
             'EN.ATM.CO2E.PC',   # Emissão de CO2 (toneladas per capita)
             'EG.ELC.ACCS.ZS',   # Acesso à eletricidade (% da população)
             'SN.ITK.DEFC.ZS']   # Prevalência de desnutrição (% da população)

# Obtendo dados
df_wb = wb.download(indicator = codigo_wb,  # Lista de indicadores
                    country = ['BR'],       # Lista de países
                    start = 2014,           # Data início AAAA
                    end = 2023)             # Data fim AAAA

# Resultado (tabela)
df_wb
```

💬 Note que a lista deste indicadores possuem valores faltantes `NaN`. Em um relatório gerencial, iremos disponibilizar somente os dados completos da seguinte forma:

```python
# Removendo dados faltantes com o
relatorio = df_wb.dropna()

# Resultado (tabela sem valores faltantes)
relatorio
```

> ⚠️ As linhas dos anos 2021 e 2022 foram excluídas!

> ⚠️ Toda a linha é excluída, incluindo as colunas que possuem informação (como na linha do ano de 2021)!

> 🗣️ Você pode substituir um `DataFrame` por ele mesmo. Aqui estamos dando nomes diferentes por fins educativos. Você pode o sobrepor dando o mesmo o nome, por exemplo: `df_wb = df_wb.dropna()`

### Selecionando linhas e colunas específicas

> 🛠️ *Exemplo prático 2:* Selecionando linhas e colunas específicas das séries sócioecômicas do Brasil

```python
# Selecionando linhas 1 e 2 e 1 e 2
relatorio.iloc[[0, 1], [0, 1]]
```

> ⚠️ Em nenhum dos casos acima, o resultado foi salvo. Somente exibido em tela!

```python
# Selecionando somente os anos pares
# Salvando o relatorio com o mesmo nome do anterior para substitui-lo
relatorio = relatorio.iloc[[0, 2, 4, 6], :]

# Visualizando resultado
relatorio
```

### Índices

🗂️ Os índices são uma série de rótulos que identificam cada linha (ou coluna) do `DataFrame`. Os rótulos podem ser números ou *strings*.

> 🛠️ *Exemplo prático 3:* Reiniciando índices das séries sócioecômicas do Brasil

```python
# Visualizando dataframe
relatorio
```

💬 Note que a lista as colunas possuem 2 linhas de nomes:

* Na linha mais baixa, temos: `country` e `year`
* Na linha mais alta, temos: `SP.DYN.LE00.IN`,	`EN.ATM.CO2E.PC`,	`EG.ELC.ACCS.ZS` e	`SN.ITK.DEFC.ZS`

💬 Além disso, perceba que o conteúdo das colunas `country` e `year` estão em negrito e das demais, não.

📝 **DICA**

*  `country` e `year` são os índices (ou rótulos) das **linhas**
*  `SP.DYN.LE00.IN`,	`EN.ATM.CO2E.PC`,	`EG.ELC.ACCS.ZS` e	`SN.ITK.DEFC.ZS` são os índices (ou rótulos) das **colunas**

👉 Portanto, `country` e `year` não contam como colunas!

Veja alguns exemplos:

```python
# Selecionando a primeira coluna
relatorio.iloc[:, [0]]
```

```python
# Selecionando a primeira coluna utilizando seu rótulo
relatorio.loc[:, ['SP.DYN.LE00.IN']]
```

```python
# Selecionando a primeira linha
relatorio.iloc[[0], :]
```

```python
# Selecionando a primeira linha utilizando seu rótulo
relatorio.loc[[('Brazil', '2020')], :]
```

> ⚠️ Neste caso, as linhas tem índices compostos que complicam bastante a manipulação! Eventualmente, vamos evitar trabalhar dessa forma.

👉 Para visualizar os índices (das linhas), podemos utilizar:

```python
relatorio.index
```

👉 Para visualizar os índices (das colunas), podemos utilizar:

```python
relatorio.columns
```

👉 Para reiniciar os índices (das linhas), podemos utilizar:

```python
relatorio.reset_index()
```

> ⚠️ Cada índice se tornou uma coluna no `DataFrame`.

👉 Agora temos o seguinte:
*   `country` é a 1a coluna (antes não contava como coluna)
*   `year` é a 2a coluna (antes não contava como coluna)
*   `SP.DYN.LE00.IN` é a 3a coluna (antes era a 1a coluna)
*   e assim sucessivamente...

> ⚠️ Não se esqueça de salvar o resultado:

```python
# Salvando o relatorio com o mesmo nome do anterior para substitui-lo
relatorio = relatorio.reset_index()

# Visualizando dataframe
relatorio
```

### Criando colunas

> 🛠️ *Exemplo prático 5:* Ações americanas em valores em reais (R\$)
*   Para um melhor entendimento dos investidores, vamos converter os valores dessas ações em dólares americanos (US\$) para reais (R$).

```python
# Carregando pacote
import yfinance as yf
```

```python
# Definindo abreviação (ticker)
ticker = ["MSFT",  # Microsoft
          "AAPL"]  # Apple

# Obtendo dados
df_acoesEUA = yf.download(ticker,                # Código das ações
                          start="2023-07-01",    # Data inicial AAAA-MM-DD
                          end="2023-07-14")      # Data final AAAA-MM-DD

# Resultado (tabela)
df_acoesEUA
```

💬 Note que, agora, o índice das colunas tem 2 níveis. Por outro lado, o índice das linhas tem somente 1.

🔄 Caso contrário do exemplo anterior.

```python
# Visualizando os índices de linha
df_acoesEUA.index
```

```python
# Visualizando os índices de linha
df_acoesEUA.columns
```

🧹 Vamos organizar este `DataFrame` com o que aprendemos até agora:
1.    Somente os valores *Adj Close*`*` (fechamento ajustado) nos interessam.
2.    Queremos que a data seja uma coluna (e não o índice de linha)
3.    Vamos renomear as colunas

> `*`📖 O fechamento ajustado é o preço de fechamento de uma ação após os ajustes para todas as distribuições de dividendos e ações corporativas que ocorreram em qualquer momento antes da abertura do dia seguinte.

```python
# TAREFA 1: Somente os valores *Adj Close* nos interessam

# Selecionando somente as 2 primeiras colunas (Adj Close)
relat_acoes = df_acoesEUA.iloc[:, [0, 1]]
```

```python
# TAREFA 2: Queremos que a data seja uma coluna

# Reiniciando índice
relat_acoes = relat_acoes.reset_index()
```

```python
# TAREFA 3: Vamos renomear as colunas

# Renomeando colunas
relat_acoes.columns = ['Data', 'Apple', 'Microsoft']
```

```python
# Visualizando o relatório
relat_acoes
```

👉 Agora, vamos criar uma nova coluna com a cotação das ações em reais (R$):[texto do link](https://)

```python
# Cotação do Dólar para Real
cotacao = 5.43   # > Ou seja, US$ 1,00 vale R$ 5,43 (suposição)

# Criando novas colunas
relat_acoes["Apple_Reais"] = relat_acoes["Apple"] * cotacao
relat_acoes["Microsoft_Reais"] = relat_acoes["Microsoft"] * cotacao
```

```python
# Visualizando o relatório
relat_acoes
```

### Filtrando colunas

> 🛠️ *Exemplo prático 7:* Filtrando a lista de ações americanas em valores em reais (R\$)
*   Queremos filtrar as ações:
    - Datas a partir de `06/07/2023`
    - Valores de `Microsoft > 340`

👉 Vamos dividir em 2 etapas:

```python
# Selecionando datas a partir de `06/07/2023`
# > Lembre-se que as datas estão no padrão americano 'AAAA-MM-DD'
relat_acoes[relat_acoes["Data"] >= '2023-07-06']

# Salvando o resultado
relat_acoes = relat_acoes[relat_acoes["Data"] >= '2023-07-06']
```

```python
# Selecionando acoes da Microsoft maiores que 340 dolares
relat_acoes[relat_acoes["Microsoft"] > 340]

# Salvando o resultado
relat_acoes = relat_acoes[relat_acoes["Microsoft"] > 340]
```

```python
# Visualizando o relatório
relat_acoes
```
